```python
import pandas as pd
```

```python
s = pd.Series([1, 2, 3, 4, 5])
s
```

```python
s.name = "My Numbers"
s.index = ['a', 'b', 'c', 'd', 'e']
s
```

```python
# indexing
s[0]
```

```python
s[1:5] #start(inclusive):end(exclusive)
```

```python
#iloc - location based indexing
s.iloc[1]
```

```python
# multiple values through iloc
s.iloc[[1,3,4]]
```
